```python
import matplotlib.pyplot as plt
import numpy as np
import math

theta=np.radians(45)

xy = np.matrix([[2,0], [4,2], [2,4], [2,0]])
ROT= np.matrix([[np.cos(theta), np.sin(theta)], \
                [-np.sin(theta), np.cos(theta)]])
REF_V = np.matrix([[1, 0] , [0,-1]])
REF_H = np.matrix([[-1, 0] , [0,1]])
REF_VH = np.matrix([[-1, 0, 0] , [0,-1, 0], [0, 0, 1]])

xy3 = np.matrix([[2,0,1], [4,2,1], [2,4,1], [2,0,1]])

plt.plot(xy[:,0], xy[:,1])

xy_ROT = xy * ROT
xy_V = xy * REF_V
xy_H = xy * REF_H
xy_VH = xy3 * REF_VH

plt.plot(xy_ROT[:,0], xy_ROT[:,1])
plt.plot(xy_V[:,0], xy_V[:,1])
plt.plot(xy_H[:,0], xy_H[:,1])
plt.plot(xy_VH[:,0], xy_VH[:,1])

plt.show()
```

```python
def rotate(origin, point, angle):    
    ox, oy = origin
    px, py = point
    qx = ox + math.cos(angle) * (px - ox) - \
                math.sin(angle) * (py - oy)
    qy = oy + math.sin(angle) * (px - ox) + \
                math.cos(angle) * (py - oy)
    return qx, qy

origin = (0, 0)
point = (5, 0)

for i in range(360):
    p = rotate(origin, point, math.radians(1))
    plt.plot([0, p[0]], [0, p[1]])
    point = (p[0], p[1])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Ellipse

delta = 45.0  # degrees

angles = np.arange(0, 360 + delta, delta)
ells = [Ellipse((1, 1), 4, 2, a) for a in angles]

a = plt.subplot(111, aspect='equal')

for e in ells:
    e.set_clip_box(a.bbox)
    e.set_alpha(0.1)
    a.add_artist(e)

plt.xlim(-2, 4)
plt.ylim(-1, 3)

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

xy = np.matrix([[0,0], [50,50], [100,0], [0,0]])*100

for i in range(5):
    xy = xy*.5
    xy = np.tile(xy, (3,1,1))
    temp = int(np.shape(xy)[0]/3)
    xy[temp:temp*2,0] = xy[temp:temp*2,0]+25*100
    xy[temp:temp*2,1] = xy[temp:temp*2,1]+25*100
    xy[temp*2:temp*3,0] = xy[temp*2:temp*3,0]+50*100

for i in range(np.shape(xy)[0]):
    plt.plot(xy[i*4:(i + 1)*4,0], xy[i*4:(i + 1)*4,1], color="black")
plt.show()
plt.savefig("triangle.pdf")
plt.savefig("triangle.png")
```
